### NASA Mars News

```python
#url link
url = 'https://redplanetscience.com'
browser.visit(url)
html=browser.html
soup=bs(html,'html.parser')
```

```python
#print pretty format and identify the structure of website
print(soup.prettify())
```

```python
# #get content title, trying to test each part 
# content_title = soup.find_all('div', class_='content_title')
# #count num of titles and print first title
# len(content_title)
# content_title[0]
```

```python
#get results as require in readme.md
content_title = soup.find_all('div', class_='content_title')
article_teaser_body= soup.find_all('div', class_='article_teaser_body')
news_t = content_title[0].text.strip()
news_p = article_teaser_body[0].text.strip()
print('NEWS_T =',news_t)
print('NEWS_P =',news_p)
```

```python
#why not to get all content titles and articles
results = soup.find_all('div', class_='list_text')

# Loop through returned results
for result in results:
    # Error handling
    try:
        # Identify and return title of listing
        content_title = result.find('div', class_='content_title').text.strip()
        
        # Identify and return article of listing
        article_teaser_body = result.find('div', class_='article_teaser_body').text.strip()

        # Run only if title and article are available
        if (content_title and article_teaser_body):
            # Print results
            print('-------------')
            print('NEWS_T = ',content_title)
            print('NEWS_P = ',article_teaser_body)

    except Exception as e:
        print(e)
```

```python
# using for loop and append
url = 'https://marshemispheres.com/'
browser.visit(url)
html = browser.html
soup = bs(html, 'html.parser')

results = soup.find_all('div', class_='description')

hemisphere_image_urls = []
for result in results:
    title = result.find('h3').text
    hemisphere_url = 'https://marshemispheres.com/' + result.a['href']
    
    browser.visit(hemisphere_url)
    html = browser.html
    soup = bs(html, 'html.parser')
    
    #click on the link to be able to identify wide-image
    hemisphere_image_url = 'https://marshemispheres.com/' + soup.find('img', class_='wide-image')['src']
    hemisphere_image_urls.append({'title': title, 'img_url': hemisphere_image_url})
hemisphere_image_urls
```
